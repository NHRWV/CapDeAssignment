```python
import csv

contents = []
categories = []
with open('petition_sample.csv', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        
        if row: #그 줄에 내용이 있는 경우에만 실시.
            contents.append(row[0]) 
            categories.append(row[1]) 
```

```python
from konlpy.tag import Okt 
twitter_tag = Okt()

from sklearn.feature_extraction.text import CountVectorizer

def tokenizer(str):
    return [token for token in twitter_tag.nouns(str) if len(token) > 1]

vec = CountVectorizer(tokenizer=tokenizer, #우선은 명사만 사용
                      decode_error ='ignore', 
                      max_df=0.3,
                      max_features = 1000) # 일단 1000개만 해본다.
contents_cv = vec.fit_transform(contents)
```

```python
contents_cv.shape
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components = 10, # 10가지 토픽 추출할 계획이다.
                                max_iter=25,
                                topic_word_prior= 0.1, doc_topic_prior=1.0,
                                learning_method='batch',
                                n_jobs= -1, # 1은 모든 프로세서를 사용한다는 뜻이라고 함.
                                random_state=0)

contents_topics = lda.fit_transform(contents_cv)

print(lda.components_.shape)
print(contents_topics.shape)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d: " % topic_idx, end='')
        print(", ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
        # 위 slicing에서 맨 뒤 -1은 역순을 의미, 그러니까 역순으로 했을 때 처음부터 n_top_words까지.
        #argsort()는 np의 함수인데 그 뭐냐 작은값부터 순서대로 나열하는 것. 그래서 그 역순으로 코드를 짠 것!!
    print()

print_top_words(lda, vec.get_feature_names(), 10)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)
kmeans.fit(lda.components_)

print('topic-word distribution dimension:', lda.components_.shape)
print('Cluster label:', kmeans.labels_)
```

```python
from sklearn.manifold import TSNE

# n_components은 임베딩 공간의 차원. 시각화하기 위해 2로 설정한다고 함...?
tsne = TSNE(n_components = 2, random_state = 42)
tsne_review = tsne.fit_transform(lda.components_)


print('TSNE dimension:', tsne_review.shape)
tsne_review[:, 0]
```
